<strong><u>Conceptos</a>:</u></strong><br>
<b><i>Deep Feature Synthesis</i></b> (<b><i>DFS</i></b>): es un método automatizado para realizar <b><i>AutoFE</i></b> en un dataset.</p>
<p style="font-size:14px;color:#34656d;font-family:'Comic Sans MS';text-align:right;border-radius:5px 5px;">
<i>Fuente:&nbsp;&nbsp;<a href="https://featuretools.alteryx.com/en/stable/getting_started/afe.html" target=\"_blank\">Deep Feature Synthesis</a></i></p> 

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Un científico de datos escribiría código para agregar datos y aplicaría diferentes funciones estadísticas.<BR>
<b><i>DFS</i></b> genera múltiples operaciones de transformación y agregación para crear características a partir del dataframe.</p>
<p style="font-size:14px;color:#34656d;font-family:'Comic Sans MS';text-align:right;border-radius:5px 5px;">

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Generación de un DFS</strong></p> 
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Del conjunto de datos <b>es</b>, se selecciona la tabla <b>customers</b> y se agrega un contador que se aplica sobre el <b>customer_id</b> (índice de tabla) y para ser contadas las sesiones en la tabla <b>sessions</b>.<br>
Para todos los campos que sean del tipo date (<b>date_of_birth - join_date</b>), se extrae el mes y se generan nuevas características.<BR>
El parametro <b>max_depth</b> con valor 1 significa que se aplica a la tabla <b>customers</b> y la tabla <b>sessions</b>
</p> 

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_entity="customers",
                                      agg_primitives=["count"],
                                      trans_primitives=["month"],
                                      max_depth=1)
```

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
A que campos de las tablas fue aplicada el <b><i>AutoFE</i></b>, puede ser consultado mediante:</p> 

```python
feature_defs
```

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
El resultado de las operaciones aplicadas sobre la tabla <b>customers</b>, se puedne consultar en el nuevo dataset creado:</p> 

```python
feature_matrix
```

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Lo que se obtiene en el nuevo dataset llamado <b>feature_matrix</b> con índice <b>customer_id</b> el <b>campo zip_code</b> al que no se le ha aplicado ninguna operación, a <b>sessions</b> se le aplico el <b>COUNT</b> y a <b>date_of_birth</b> y <b>join_date</b> se les aplicó <b>MONTH</b><BR><BR>
<b><i>Primitiva de agregación</i></b>: calcula un valor único basado en muchas sesiones relacionadas<BR>
<b><i>Primitiva de transformación</i></b>: toma un valor y lo transforme en otro</p> 
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:center;border-radius:5px 5px;">
<strong>_____________________________________________________________</strong></p>     

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<strong>Generar un DFS con profundidad 2</strong>:<BR>
El parametro <b>max_depth</b> con valor 1 significa que se aplica a la tabla <b>customers</b>, la tabla <b>sessions</b>  y la tabla <b>transactions</b></p> 

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_entity="customers",
                                      agg_primitives=["mean", "sum", "mode"],
                                      trans_primitives=["month", "hour"],
                                      max_depth=2)
```

```python
feature_defs
```

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
El resultado de las operaciones aplicadas sobre la tabla <b>customers</b>, se pueden consultar en el nuevo dataset creado:</p> 

```python
feature_matrix
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Al nuevo dataframe <b>feature_matrix</b>, se pueden aplicar diversas operaciones propias de <b>Pandas</b></p> 

```python
feature_matrix[['MEAN(sessions.SUM(transactions.amount))']].sort_values(by=['customer_id'])
```

```python
feature_matrix[['MODE(sessions.HOUR(session_start))']].sort_values(by=['customer_id'])
```

<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
Las <b><i>Feature Primitives</i></b> permiten definir también eventos de tiempo</b>, como sera <b><i>mean</i></b> <b><i>max</i></b> - <b><i>min</i></b> - <b><i>std</i></b> - <b><i>skew</i> - <b><i>time_since_previous</i></b></p> 

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_entity="customers",
                                      agg_primitives=["mean", "max", "min", "std", "skew"],
                                      trans_primitives=["time_since_previous"],)
```

<b> Aggregation Primitive vs Transform Primitive</b></p>

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<b><i>Aggregation Primitive</i></b>: Estas primitivas toman diversas instancias como entrada y entregan un valor único</p> 

![AG](./img/02.png)

<a id = "1.0"></a>
<p style="font-size:15px;color:#34656d;font-family:'Comic Sans MS';text-align:left;border-radius:5px 5px;">
<b><i>Trasform Primitive</i></b>: Estas primitivas toman una o más variables de una entidad como entrada y generan una nueva variable para esa entidad</p>

## Reemplazo de nulos

Aca se toma el promedio para los floats, la mediana para los int y la moda para las categorías

```python
int64_columns = df.select_dtypes(include=['Int64']).columns.tolist()
float_columns = df.select_dtypes(include=['float64']).columns.tolist()
categorical_columns = df.select_dtypes(include=['category']).columns.tolist()
boolean_columns = df.select_dtypes(include=['bool']).columns.tolist()
```

```python
df[categorical_columns].isnull().sum(axis=0).sort_values(ascending=False)
```

```python
for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
df[categorical_columns].isnull().sum(axis=0).sort_values(ascending=False)
```

```python
df[float_columns].isnull().sum(axis=0).sort_values(ascending=False)
```

```python
for col in float_columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df[float_columns].isnull().sum(axis=0).sort_values(ascending=False)
```

```python
df[int64_columns].isnull().sum(axis=0).sort_values(ascending=False)
```

```python
for col in int64_columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
df[int64_columns].isnull().sum(axis=0).sort_values(ascending=False)
```
